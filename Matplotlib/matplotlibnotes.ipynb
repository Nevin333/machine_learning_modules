```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('day')
plt.ylabel('Temperature')
plt.title('Weather Info')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='Avg')
plt.legend(shadow=True)
plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("majo0r companies")
plt.ylabel("Revenue")
plt.title("Revenue model for top companies")
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xlabel("majo0r companies")
plt.ylabel("Revenue")
plt.title("Revenue model for top companies")
plt.bar(xpos-0.3,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.xticks(xpos,company)
plt.legend()
plt.show()
```

```python
np.random.seed(10)
data_1=np.random.normal(100,10,200)
data_2=np.random.normal(90,20,200)
data_3=np.random.normal(80,30,200)
data_4=np.random.normal(70,40,200)

data=[data_1,data_2,data_3,data_4]
plt.figure(figsize=(10,5))

plt.boxplot(data)
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%",shadow=True,explode=[0.1,0.1,0,0,0],startangle=65)
plt.legend()
plt.savefig("piechart.jpg")
```
